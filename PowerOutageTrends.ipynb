## Check the Min / Max / Average Values of the Data

3. Variable List: 
 - Fips_code: Code for the county
 - County: County name
 - State: State (includes Puerto Rico and District of Columbia)
 - Sum: Total number of customers without power (may be one building, one meter)
 - Run_start_time: “MM/DD/YY 00:00” at 15-minute intervals


```python
poweroutage2022.head()
```

```python
import matplotlib.pyplot as plt 

statedata = poweroutage2022.groupby("state").sum()["sum"]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
statedata.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Total Customers Without Power')
plt.title('Power Outages by State in 2022')
plt.xticks(rotation=90)  # Rotating x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
florida_data = poweroutage2022[poweroutage2022['state'] == 'Florida']
florida_data.head()
```

```python
floridaCounties = florida_data.groupby("county").sum()["sum"]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
floridaCounties.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Total Customers Without Power')
plt.title('Power Outages in Florida in 2022')
plt.xticks(rotation=90)  # Rotating x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
